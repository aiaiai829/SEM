```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Search terms report.csv',delimiter='\t',encoding = 'utf-16',engine='python', skiprows=2, thousands=',') 
df=df.dropna()
```

```python
df
```

```python
# Remove any unwanted characters from the 'Keyword' column in both dataframes
df['Keyword'] = df['Keyword'].str.replace('"', '')
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Concatenate the two columns to fit the vectorizer
texts = pd.concat([df['Search term'], df['Keyword']])
tfidf_matrix = vectorizer.fit_transform(texts)

# Split the TF-IDF matrix into two: one for 'Search term' and one for 'Keywords'
st_matrix = tfidf_matrix[:len(df['Search term'])]
kw_matrix = tfidf_matrix[len(df['Search term']):]

# Calculate cosine similarity
similarities = [cosine_similarity(st_matrix[i], kw_matrix[i])[0][0] for i in range(len(df))]

# Add the similarity scores to the dataframe
df['similarity'] = similarities
```

```python
from gensim.models import KeyedVectors
import gensim.downloader as api
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load pre-trained Word2Vec embeddings
model = api.load("word2vec-google-news-300")
```

```python
def get_vector(sentence):
    """Get the average vector for the sentence."""
    words = sentence.split()
    vectors = [model[word] for word in words if word in model.key_to_index]
    if vectors:
        return sum(vectors) / len(vectors)
    else:
        return [0]*300

def calculate_similarity(row):
    """Calculate cosine similarity between search term and keyword."""
    st_vector = get_vector(row['Search term'])
    kw_vector = get_vector(row['Keyword'])
    return cosine_similarity([st_vector], [kw_vector])[0][0]

# Calculate similarity scores
df['embedding_similarity'] = df.apply(calculate_similarity, axis=1)
```

```python
df
```

```python
# Create df_similarity
df_similarity = df[['Search term', 'Keyword', 'Match type','similarity','embedding_similarity', 'Ad group','Impr.', 'Clicks']].copy()
df_similarity 
```

```python
# Group by 'Keyword', 'Match type', 'Ad group' and aggregate the desired columns
df_grouped = df_similarity.groupby(['Keyword', 'Match type', 'Ad group']).agg({
    'similarity': 'mean',
    'embedding_similarity': 'mean',
    'Impr.': 'sum',
    'Clicks': 'sum'
}).reset_index()

# Create the 'click/imp' column
df_grouped['click/imp'] = df_grouped['Clicks'] / df_grouped['Impr.']

# Handle potential infinite values (if Impr. is 0)
df_grouped['click/imp'].replace([float('inf'), -float('inf')], 0, inplace=True)

# Sort the results by 'click/imp'
df_grouped = df_grouped.sort_values(by='click/imp', ascending=False)

```

```python
import duckdb
import ipywidgets as widgets
from IPython.display import display, clear_output,HTML
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
# Function to display visualizations based on selections
def display_visualizations(ad_group, keyword, match_type):
    if ad_group == "all" and keyword == "all":
        filtered_df = df
    elif ad_group == "all":
        filtered_df = df[df['Keyword'] == keyword]
    elif keyword == "all":
        filtered_df = df[df['Ad group'] == ad_group]
    else:
        filtered_df = df[(df['Keyword'] == keyword) & (df['Ad group'] == ad_group)]
    
    # Filter by match type if not "all"
    if match_type != "all":
        filtered_df = filtered_df[filtered_df['Match type'] == match_type]
    
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))
    
    # Check if there's data for the selected values
    if len(filtered_df) == 0:
        axs[0].text(0.5, 0.5, 'Not enough data', ha='center', va='center', fontsize=20)
        axs[0].axis('off')
        axs[1].text(0.5, 0.5, 'Not enough data', ha='center', va='center', fontsize=20)
        axs[1].axis('off')
    else:
        text = ' '.join(filtered_df['Search term'])
    
        # Display word cloud
        wordcloud = WordCloud(width=400, height=400, background_color='white').generate(text)
        axs[0].imshow(wordcloud, interpolation='bilinear')
        axs[0].axis('off')
        axs[0].set_title('Search Term Word Cloud')

        # Display bar plot
        metrics = [ 'Impr.','Clicks',  'Conversions', 'Cost','similarity', 'embedding_similarity'] 
        values = []
        for metric in metrics:
            if metric in ['similarity', 'embedding_similarity']:
                values.append(filtered_df[metric].mean())  # Calculate mean for similarity metrics
            else:
                values.append(filtered_df[metric].sum())  # Calculate sum for other metrics
        axs[1].bar(metrics, values, color=['blue', 'green', 'red', 'purple', 'cyan', 'orange'])
        axs[1].set_title('Metrics Overview')
        for i, v in enumerate(values):
            axs[1].text(i, v + 0.05, str(round(v, 2)), ha='center', va='bottom', fontsize=9)
    
    plt.tight_layout()
    plt.show()

# Function to update keyword dropdown based on ad group selection
def on_ad_group_change(change):
    if change['new'] == "all":
        new_options = ['all'] + list(df['Keyword'].unique())
    else:
        new_options = ['all'] + list(df[df['Ad group'] == change['new']]['Keyword'].unique())
    keyword_dropdown.options = new_options

# Set up the Ad group dropdown with an observer to update Keyword dropdown
ad_group_dropdown = widgets.Dropdown(options=['all'] + list(df['Ad group'].unique()), description='Ad group:')
ad_group_dropdown.observe(on_ad_group_change, names='value')

# Interactive visualization setup
keyword_dropdown = widgets.Dropdown(options=['all'] + list(df['Keyword'].unique()), description='Keyword:')
match_type_dropdown = widgets.Dropdown(options=['all'] + list(df['Match type'].unique()), description='Match type:')

# Display widgets and visualization
w = widgets.interactive(display_visualizations, ad_group=ad_group_dropdown, keyword=keyword_dropdown, match_type=match_type_dropdown)
display(w)
```

```python
# Create widgets
keyword_search = widgets.Text(placeholder='Enter keyword to search', description='Keyword:')
output_table = widgets.Output()
display(keyword_search, output_table)

# Function to update the displayed table based on keyword search
def update_table(keyword):
    with output_table:
        output_table.clear_output()
        filtered_df = df_new[df_new['Search term'].str.contains(keyword, case=False, na=False)]
        if not filtered_df.empty:
            display(HTML(filtered_df[['Search term', 'Keyword', 'Ad group', 'Status', 'Cost', 'Clicks', 'Impr.', 'Conversions', 'similarity', 'embedding_similarity']].to_html(index=False)))
        else:
            print("No matching records found.")

# Define a callback for the keyword search widget
def keyword_search_callback(change):
    keyword = change.new
    update_table(keyword)

# Attach the callback to the keyword search widget
keyword_search.observe(keyword_search_callback, names='value')

# Initial display (empty table)
update_table('')
```
